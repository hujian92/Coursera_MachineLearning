### 特征映射(Feature map)<span id='fm'/>
$$mapFeature(x) = 
 \left[
 \begin{matrix}
   1 \\
   x_1 \\
   x_2 \\
   x_1^2 \\
   x_1 x_2 \\
   x_2^2 \\
   x_1^3 \\
   \vdots \\
   x_1x_2^5 \\
   x_2^6
  \end{matrix}
  \right] 
$$

```python
# 特征映射,返回映射后的特征矩阵,和特征个数
def mapFeature(x1, x2, power): 
    count = (1 + power + 1) * (power + 1) / 2
    out = np.ones((len(x1), count), dtype='float32')
    k = 0
    for i in np.arange(power) + 1:
        for j in np.arange(i + 1):
            out[:, k + 1] = (x1 ** (i - j)) * (x2 ** j)
            k += 1
    return out, count
```

```python
# mapFeature
power = 6
xmap, count = mapFeature(X[:, 0], X[:, 1], 6)
```

#### sigmod function
$$ g(z) = \frac {1} { 1 + e ^ {-z} } $$

```python
def sigmoid(z):
    return 1 / (1 + T.exp(-z))
```

### 构建模型
#### hypothesis
$$ h _ { \theta } (x) = g ( \theta ^ Tx ) $$
#### loss function
$$ J ( \theta ) = \frac {1} {m} \sum_{ i=1 } ^ {m} \left[ - y^ { (i) } \log ( h_{ \theta } ( x ^ { (i) } ) ) - (1 - y^{ (i) } ) \log ( 1 - h_{ \theta } ( x ^ { (i) } ) ) \right]  + \frac { \lambda } { 2m } \sum _ { j = 1 } ^ {n} \theta _ j^2 (正则项不惩罚\theta_0) $$

```python
x = T.matrix('x') # (118, 28)
y = T.vector('y')
theta = theano.shared(np.zeros([count, 1], dtype='float32')) # (28, 1)
alpha = T.scalar('alpha') # 学习率
lambdaReg = T.scalar('lambdaReg') # 正则项系数
pre_prob = sigmoid(x.dot(theta)) # (118, 1)
thetaExclude0 = theta[1:, :]  # 不惩罚theta[0, 0] , 还可以写成theta[1:], 或者T.set_subtensor(theta[0, 0], 0) 
loss = T.mean(-y.dimshuffle(0, 'x') * T.log(pre_prob) - (1- y.dimshuffle(0, 'x')) * T.log(1 - pre_prob)) +  lambdaReg * (thetaExclude0 ** 2).sum() / (2 * x.shape[0]) #  不对theta[0]正则化
calcLoss = theano.function([x, y, lambdaReg], loss)
```

#### gradient w.s.t theta
$$ 
\begin{align} 
    & \frac { \partial { J ( \theta ) } } { \partial { \theta_0 } } = \frac {1} {m} \sum _ { i = 1 } ^ {m} ( h _ { \theta } ( x ^ { (i) } )  - y ^ { (i) } ) x _ j ^ { (i) } \qquad & for \quad j = 0 \notag \\
    & \frac{\partial{J(\theta)}} {\partial{\theta_j}} = (\frac{1}{m} \sum_{i=1}^{m} (h_{\theta}(x^{(i)})  - y^{(i)}) x_j^{(i)}) +  \frac{\lambda}{m} \theta_j  \qquad & for \quad j \geq 0 \notag
 \end{align}
$$  

#### update theta with gradient descent
$$ \theta = \theta - \alpha \nabla \theta $$

```python
thetaTemp =  T.set_subtensor(theta[0, 0], 0) # 手动求导, theta0比较特殊,
grad_all = (x.T.dot(pre_prob - y.dimshuffle(0, 'x')) +  lambdaReg * thetaTemp) /  x.shape[0].astype('float32') # 求梯度, 使用矩阵乘法,就不用difshuffle了
gradient_step_all = theano.function([x, y, alpha, lambdaReg],updates={theta : theta - alpha * grad_all})  # 梯度下降
```

```python
grad_auto = T.grad(loss, theta) # 自动求梯度
gradient_step_auto = theano.function([x, y, alpha, lambdaReg],updates={theta : theta - alpha * grad_auto})
```

```python
print calcLoss(xmap, Y, 1)
print grad_all.eval({x:xmap, y:Y, lambdaReg:1})
print grad_auto.eval({x:xmap, y:Y, lambdaReg:1})
```

### 梯度下降法,固定学习率训练模型<span id='static'/>

```python
# 画决策边界
def plotDecisionBoundary(theta, X, Y,lambdaValue = 1, axis = None): 
    if axis == None:
        axis = plt.gca()
    # 画出数据点
    pos = axis.scatter(X[Y == 1, 0], X[Y == 1, 1], s=32, c='red', marker='x')
    neg = axis.scatter(X[Y == 0, 0], X[Y == 0, 1], s=32, c='yellow', marker='o')
    axis.set_xlabel('Microchip Test 1')
    axis.set_ylabel('Microchip Test 2')
    axis.legend((pos, neg), ('y = 1', 'y = 0'), scatterpoints=1, loc='lower left', ncol=1, fontsize=12) # 这样给scatter添加lagend
    
    # 画决策边界
    u = np.linspace(-1, 1.5, 50)
    u.shape = -1, 1 # 转成(50, 1) 为了调用mapFeature中的len(..)函数
    v = np.linspace(-1, 1.5, 50)
    v.shape = -1, 1
    a, b = np.meshgrid(u, v)
    
    z = np.zeros((len(u), len(v)))
    print z.shape
    for i in np.arange(len(u)):
        for j in np.arange(len(v)):
            z[i, j] = mapFeature(u[i], v[j], 6)[0].dot(theta)
            
    C = axis.contour(a, b, z.T, [0], colors='blue', linewidth=.5) # 只画线,并且只画出等于0的值
    # Kind of a hacky way to display a text on top of the decision boundary
    myfmt = { 0:'Lambda = %d'%lambdaValue}
    plt.clabel(C, inline=1, fontsize=15, fmt=myfmt)
    
```

```python
max_iters = 5000
lambdaInit = 1
theta.set_value(np.zeros([count, 1], dtype='float32'))
J_history = np.zeros((max_iters, 1))
for i in range(5000):
    gradient_step_auto(xmap, Y, 0.1, lambdaInit)
    J_history[i] = calcLoss(xmap, Y, lambdaInit)
plotDecisionBoundary(theta.get_value(), X, Y)
```

### 梯度下降法, 动态调整学习率训练<span id='dy'/>

```python
# 训练,动态调整学习率,loss减少时才更新theta,否则仅降低学习率(但是这个方法在这里不行,需要在loss增大时同时更新theta和学习率)
theta.set_value(np.zeros([count, 1], dtype='float32'))
max_iters = 5000
best_theta = theta.get_value()
alphaValue = 0.1
best_loss = np.Inf
lambdaInit = 1
J_history = np.zeros((max_iters, 1))
for i in range(max_iters):
    theta.set_value(best_theta)
    gradient_step_all(xmap, Y, alphaValue, lambdaInit)
    tempLoss = calcLoss(xmap, Y, lambdaInit)
    J_history[i] = tempLoss
    if tempLoss < best_loss: #  loss迭代时不会进入if,即不会更新theta而是一直更新alpha
        best_loss = tempLoss
        best_theta = theta.get_value()
    else:
        alphaValue = alphaValue * 0.999
        best_theta = theta.get_value()   # 要在loss增大时也更新theta才有效果
        
fig, axes = plt.subplots(1, 2, figsize=(15, 7))
plotDecisionBoundary(thetaValue.reshape(28,1), X, Y, axis=axes[0])        

# Plot the convergence graph
plt.sca(axes[1])
plt.plot(np.arange(max_iters) + 1,J_history)
plt.xlabel('Number of iterations')
plt.ylabel('Cost J')
```

### 使用scipy.optimize优化,并动态演示迭代过程<span id='scipy'/>

```python
# 将matplotlib动画嵌入为HTML视频
from matplotlib import animation, pyplot as plt
from tempfile import NamedTemporaryFile
from IPython.display import HTML

VIDEO_TAG = """<video controls autoplay>
 <source src="data:{0}">
 Your browser does not support the video tag.
</video>"""

def anim_to_html(anim, fps = 20):
    if not hasattr(anim, '_encoded_video'):
        with NamedTemporaryFile(suffix='.m4v') as f:
            anim.save(f.name, fps=fps, extra_args=['-vcodec', 'libx264'])
            video = open(f.name, "rb").read()
        anim._encoded_video = 'video/mp4;base64,' + video.encode("base64")
    # prevent figure displayed as a PNG below the animation
    #plt.close()
    return VIDEO_TAG.format(anim._encoded_video)

def display_animation(anim, fps = 20):
    #plt.close(anim._fig)
    return HTML(anim_to_html(anim, fps = fps))
#animation.Animation._repr_html_ = anim_to_html
```

```python
from scipy import optimize

# 计算指定theta时的cost,使用optimize的函数时,传入的待优化参数是一个一维向量,所以要转成二维列向量,并且使用fmin时还需要转成float32类型
def computeCost(theta_value, lambda_value):
    theta.set_value(theta_value.reshape(28,1).astype('float32'))
    return calcLoss(xmap, Y, lambda_value)
# 计算指定theta时的梯度
calcGrad= theano.function([x, y, lambdaReg], grad_auto)
def train_fn_grad(theta_value, lambda_value):
    theta.set_value(theta_value.reshape(28,1).astype('float32'))
    return np.array(calcGrad(xmap, Y, lambda_value).reshape(28)) # 梯度也要转为一维向量


def fitModel(lambdaValue = 1):
    # 使用共轭梯度法优化
    thetaV = [] # 存储每一次迭代后的结果
    thetaValue = optimize.fmin_cg( # or fmin_bfgs
        f=computeCost,
        x0=np.random.rand(28,1),
        args=(lambdaValue, ), # 一个元素的tuple
        fprime=train_fn_grad,
        disp=1,
        maxiter=1000,
        callback=lambda t : thetaV.append(t.reshape(28, 1))
    )
    return thetaValue, thetaV

thetaValue, thetaV = fitModel(lambdaValue = 1)
# 动态演示迭代过程
fig = plt.figure()

u = np.linspace(-1, 1.5, 50)
u.shape = -1, 1 # 转成(50, 1) 为了调用mapFeature中的len(..)函数
v = np.linspace(-1, 1.5, 50)
v.shape = -1, 1
a, b = np.meshgrid(u, v)
z = np.zeros((len(u), len(v)))

def plotDecisionBoundaryAnimation(k): # 画出决策边界
    plt.gca().clear()
    # 画出数据点
    pos = plt.scatter(X[Y == 1, 0], X[Y == 1, 1], s=32, c='red', marker='x')
    neg = plt.scatter(X[Y == 0, 0], X[Y == 0, 1], s=32, c='yellow', marker='o')
    plt.text(-1, 1.1, 'frame: %d' % k)
    plt.xlabel('Microchip Test 1')
    plt.ylabel('Microchip Test 2')
    plt.legend((pos, neg), ('y = 1', 'y = 0'), scatterpoints=1, loc='lower left', ncol=1, fontsize=12) # 给scatter添加lagend
    
    for (i, j), _ in np.ndenumerate(z):
        z[i, j] = mapFeature([a[i, j]], [b[i, j]], 6)[0].dot(thetaV[k])

    C = plt.contour(a, b, z, [0], colors='blue', linewidth=.5) # 只画线,并且只画出等于0的值

#  动画演示  
anim = animation.FuncAnimation(fig, plotDecisionBoundaryAnimation, frames=len(thetaV), repeat=False) # 每10ms调用一次animate绘图
display_animation(anim, 1) # 视频动画演示
```

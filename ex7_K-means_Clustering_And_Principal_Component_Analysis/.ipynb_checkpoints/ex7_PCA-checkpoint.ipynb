```python
import numpy as np
import theano
import theano.tensor as T
import scipy.io as sio
from scipy import misc
import matplotlib.pyplot as plt
import matplotlib.cm as cm
from matplotlib.path import Path
import matplotlib.patches as patches
from mpl_toolkits.mplot3d import Axes3D

%matplotlib inline
```

### 读取数据

```python
data = sio.loadmat('data/ex7data1.mat')
X = data['X'].astype('float32')
plt.scatter(X[:, 0], X[:, 1], s=30, marker=u'o', edgecolors='b', facecolors='none') # visualize data 
```

```python
def featureNormalize(x): # 特征尺度变换
    mu = x.mean(0)
    #sigma = x.std(0) # 估计的是总体标注差,除以的是m, matlab是m - 1
    sigma = np.sqrt(((x - mu) ** 2).sum(0) / (len(x) - 1)) # 样本标准差
    x_norm = (x - mu) / sigma
    return x_norm, mu, sigma
```

```python
X_norm, mu, sigma = featureNormalize(X)
```

### 构建模型

```python
x = theano.shared(X_norm, name='x') # (50, 2)
k = T.iscalar('k') # reduce dim to k

covariance_mat = T.dot(x.T, x) / x.shape[0] 
U, V, D = T.nlinalg.svd(covariance_mat)  # singular value decomposition
svd = theano.function([], [U, V])

U_reduce = U[:, :k] # the first k principal components
Z = T.dot(x, U_reduce) # project data to lower dim (50 ,1)
x_rec = T.dot(Z, U_reduce.T) #  recover data from lower dim  (50, 2)
projectData = theano.function([k], Z)
recoverDate = theano.function([k], x_rec)
```

```python
u, v = svd()
```

### 画出特征向量的方向<span id='dr'/>

```python
plt.figure(figsize=(8,8))
plt.scatter(X[:, 0], X[:, 1], s=30, marker=u'o', edgecolors='b', facecolors='none') # visualize data 
plt.scatter(mu[0], mu[1], s=50, marker=u'x', c='r') # plot mean point
plt.gca().set_aspect('equal')

plt.quiver(mu[0], mu[1], u[0, 0], u[1 ,0], scale=8 * v[1], color='r')
plt.quiver(mu[0], mu[1], u[1, 0], u[1 ,1], scale=8 * v[0], color='r')
```

### 数据降维<span id='drr'/>

```python
z = projectData(1)
X_rec = recoverDate(1)
```

```python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111)

verts = []
codes = []
for i in range(X_norm.shape[0]): # draw line between original data and recover data
    verts = verts + [X_norm[i], X_rec[i]]
    codes = codes + [Path.MOVETO, Path.LINETO]
    
path = Path(verts, codes)
patch = patches.PathPatch(path, facecolor='orange', lw=1)
ax.add_patch(patch)
ax.scatter(X_rec[:, 0], X_rec[:, 1], s=30, marker=u'o', edgecolors='r', facecolors='none') # visualize  recoverd data 
ax.scatter(X_norm[:, 0], X_norm[:, 1], s=30, marker=u'o', edgecolors='b', facecolors='none') # visualize original data 
plt.scatter(-1 * z[:, 0], -3 * np.ones(50), s=30, marker=u'o', edgecolors='r', facecolors='none') # visualize  recoverd data 
ax.set_xlim(-4, 3)
ax.set_ylim(-4, 3)
```

## PCA on Face Data: Eigenfaces<span id='face'/>

```python
# 画出数字
def displayData(X, example_width=-1):
    if example_width == -1:
        example_width = np.round(np.sqrt(X.shape[1])).astype('int8')
    # 计算每个小图的宽高
    m, n = X.shape
    example_height = (n / example_width)
    
    # 计算整个大拼图的宽高
    display_rows = np.floor(np.sqrt(m))
    display_cols = np.ceil(m / display_rows)
    
    #  小图之间的间隙
    pad = 1

    # 初始化大图矩阵
    display_array = -np.ones([pad + display_rows * (example_height + pad), pad + display_cols * (example_width + pad)])
    # 装填图像矩阵
    curr_ex = 0
    for j in np.arange(display_rows):
        for i in np.arange(display_cols):
            if curr_ex >= m:
                break
            max_val = np.max(np.abs(X[curr_ex, :]))
            offset_row = int(pad + j * (example_height + pad))
            offset_col = int(pad + i * (example_width + pad))
            display_array[offset_row : offset_row + example_height,
                                     offset_col : offset_col + example_width] = np.reshape(X[curr_ex, :], [example_width, example_height], order='F') / max_val # 按照Fortran顺序reshape(matlab也是按F顺序),或者按照默认的C顺序,然后转置
            curr_ex += 1
        if curr_ex >= m:
            break
    plt.imshow(display_array, cmap=plt.cm.binary_r)
```

```python
# load face images data
data = sio.loadmat('data/ex7faces.mat')
X = data['X'].astype('float32')
displayData(X[:100,:])
```

```python
X_norm, mu, sigma = featureNormalize(X)
x.set_value(X_norm)
```

### 画出特征脸(eigenface)<span id='eigenface'/>

```python
u, v = svd() 
displayData(u[:, :36].T) # Visualize the top 36 eigenvectors found
```

求出降维的投影平面方程: $ax+by+cz+d=0$   
求解条件: 点mu在平面上, 平面上两条直线的方向向量为u[0], u[1] ...

```python
a = -2.0459
b = 2.8974
c = -1.0
d = 0.1105
```

### 画出降维的投影平面和原始数据

```python
fig = plt.figure()
ax3D = fig.add_subplot(111, projection = '3d') 

points_random = X_norm[sel]
xa = np.arange(points_random[:, 0].min() ,points_random[:, 0].max() , 0.1)
ya = np.arange(points_random[:, 1].min() ,points_random[:, 1].max() , 0.1)
xa, ya = np.meshgrid(xa, ya)
z = a * xa + b * ya + d # compute z values

ax3D.plot_surface(xa, ya,  z, rstride=1, cstride=1, alpha=0.2, cmap=plt.cm.jet, shade=False)  # plot projection plane
ax3D.scatter(points_random[:, 0], points_random[:, 1], points_random[:,2], c = colors[idx_[sel]], s=20, marker=u'.')
```

### 对比降维后复原的数据和投影数据

```python
fig = plt.figure()
ax3D = fig.add_subplot(111, projection = '3d') 

points_random = X_rec[sel]
xa = np.arange(points_random[:, 0].min() ,points_random[:, 0].max() , 0.1)
ya = np.arange(points_random[:, 1].min() ,points_random[:, 1].max() , 0.1)
xa, ya = np.meshgrid(xa, ya)
z = a * xa + b * ya + d

point2 = X_norm[sel]

for p1, p2 in zip(point2, points_random): # draw line between original data and recover data
    ax3D.plot([p1[0], p2[0]], [p1[1], p2[1]], [p1[2], p2[2]], 'k-')

# project original data onto projection plane
normVec = np.array([[a],  [b], [c]]) 
k = -(np.dot(X_norm, normVec) + c) / np.sum(normVec ** 2)
X_pro = np.dot(k, normVec.T) + X_norm 
point3 = X_pro[sel]

ax3D.plot_surface(xa, ya,  z, rstride=1, cstride=1, alpha=0.2, cmap=plt.cm.jet, shade=False) 
scatter_rec = ax3D.scatter(points_random[:, 0], points_random[:, 1], points_random[:,2], c = 'g', s=30, marker=u'o')
scatter_ori = ax3D.scatter(point2[:, 0], point2[:, 1], point2[:,2], c = 'r', s=30, marker=u'o')
scatter_pro = ax3D.scatter(point3[:, 0], point3[:, 1], point3[:,2], c = 'b', s=30, marker=u's')
ax3D.legend([scatter_rec, scatter_ori, scatter_pro], ['recovered data', 'original data', 'projection data'])
```
